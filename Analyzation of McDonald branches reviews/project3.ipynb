```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
plt.style.use('ggplot')
```

```python
df = pd.read_csv('McDonald_s_Reviews.csv', encoding='latin-1')
```

```python
df = df[['reviewer_id', 'store_address',
       'rating_count', 'review_time', 'review', 'rating']]
```

```python
for i in df['review'] :
    if i == None:
        print('1')
```

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['review']
    myId = row['review']
    res[myId] = sia.polarity_scores(text)
```

```python
vaders = pd.DataFrame(res).T
vaders = vaders.reset_index().rename(columns={'index': 'Id'})
# vaders = 
```

```python
vaders.head()
```

```python
df.head()
```

```python
vaders = pd.concat([df, vaders], axis=1, join='inner')
```

```python
vaders.head()
```

```python
vaders[['reviewer_id', 'store_address', 'rating_count', 'review_time', 'review',
       'rating', 'neg', 'neu', 'pos', 'compound']]
```

```python
vaders = vaders.loc[~(vaders['neu']==1)]
```

```python
vaders.shape
```

```python
store_address = []
```

```python
for i in vaders['store_address']:
    if i in store_address:
        continue
    store_address.append(i)
```

```python
print(len(store_address))
```

```python
noneData = []
for i in range(17127):
    if i in vaders['neg']:
        continue
    noneData.append(i)
```

```python
myDfData = []
for i in store_address:
    ave_scores = 0
    ave_neg = 0
    ave_neu = 0
    ave_pos = 0
    ave_compound = 0
    shomarande = 0
    counts = 0
    for j in vaders['store_address']:
        if shomarande in noneData :
            shomarande += 1
            continue
        if i == j:
            # ave_scores += int(vaders['rating'][shomarande])
            ave_neg += float(vaders['neg'][shomarande])
            ave_neu += float(vaders['neu'][shomarande])
            ave_pos += float(vaders['pos'][shomarande])
            ave_compound += float(vaders['compound'][shomarande])
            counts += 1
        shomarande += 1
    myDfData.append((i, ave_neg/counts, ave_neu/counts, ave_pos/counts, ave_compound/counts))
```

```python
myDataFrame = pd.DataFrame(myDfData, columns=['Store_address', 'Ave_Neg',
                                              'Ave_Neu', 'Ave_Pos', 'Ave_Compound'])
                                              
```

```python
myDataFrame
```

```python
rating = []
for i in range(1, 32):
    rating.append(i)
Rating = pd.DataFrame(tuple(rating), columns=['Rating'])
```

```python
Rating
```

```python
result = myDataFrame.sort_values('Ave_Pos', ascending=False).reset_index()
```

```python
result = result[['Store_address', 'Ave_Neg', 'Ave_Neu', 'Ave_Pos',
       'Ave_Compound']]
```

```python
Result = pd.concat([Rating, result], axis=1, join='inner')
```
